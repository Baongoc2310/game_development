```python
gamescopy.columns = ['ID','Name','Subtitle','Average_User_Rating','User_Rating_Count','Price','In_App_Purchases','Description','Developer','Age_Rating','Language','Size','Primary_Genre','Genres','Original_Release_Date','Curren_Version_Release_Date']
```

Nếu một game không có đánh giá từ người chơi, ta không thể biết chất lượng của nó, nên loại bỏ để tránh làm sai lệch kết quả phân tích.

```python
gamescopy = gamescopy[pd.notnull(gamescopy['Average_User_Rating'])]
```

```python
import datetime
gamescopy.loc[:, 'Original_Release_Date'] = pd.to_datetime(gamescopy['Original_Release_Date'], format="%d/%m/%Y")
gamescopy.loc[:, 'Curren_Version_Release_Date'] = pd.to_datetime(gamescopy['Curren_Version_Release_Date'], format="%d/%m/%Y")
```

```python
gamescopy.loc[:,'Length'] = gamescopy.Curren_Version_Release_Date - gamescopy.Original_Release_Date
```

```python
gamescopy
```

```python
Low_User_Count = gamescopy[(gamescopy['User_Rating_Count'] < 200) & 
(gamescopy['Length'] < datetime.timedelta(days =175))].index
gamescopy = gamescopy.drop(Low_User_Count)
```

Lọc ra các game có hai điều kiện sau:
- Có ít hơn 200 lượt đánh giá (User_Rating_Count < 200).
- Thời gian từ ngày phát hành đến lần cập nhật gần nhất dưới 6 tháng (175 ngày).
 Lý do :
🔹 Game có ít đánh giá (<200) thì không đủ dữ liệu để xác định chất lượng thật sự.
🔹 Nếu game mới ra mắt và cập nhật nhanh (<6 tháng), có thể những đánh giá ban đầu chưa phản ánh đúng chất lượng game (có thể là đánh giá ảo từ nhà phát triển).

```python
gamescopy = gamescopy.copy()
gamescopy['Genres'] = gamescopy['Genres'].str.replace(',', '').str.replace('Games', '').str.replace('Entertainment','').str.replace('Strategy','')
gamescopy['Genres'] = gamescopy['Genres'].str.split(' ').map(lambda x: ' '.join(sorted(x)))
gamescopy['Genres'] = gamescopy['Genres'].str.strip()
```

```python
Non_Main_Genres = gamescopy[ gamescopy['Genres'].str.contains('Puzzle') & 
    gamescopy['Genres'].str.contains('Action') & 
    gamescopy['Genres'].str.contains('Family') & 
    gamescopy['Genres'].str.contains('Education') & 
    gamescopy['Genres'].str.contains('Family') &
    gamescopy['Genres'].str.contains('Adventure') & 
    gamescopy['Genres'].str.contains('Board') & 
    gamescopy['Genres'].str.contains('Role')].index
gamescopy.drop(Non_Main_Genres, inplace = True)
```

```python
gamescopy.loc[gamescopy['Genres'].str.contains('Puzzle'),'Genres'] = 'Puzzle'
gamescopy.loc[gamescopy['Genres'].str.contains('Board'),'Genres'] = 'Puzzle'
gamescopy.loc[gamescopy['Genres'].str.contains('Action'),'Genres'] = 'Action'
gamescopy.loc[gamescopy['Genres'].str.contains('Adventure'),'Genres'] = 'Adventure'
gamescopy.loc[gamescopy['Genres'].str.contains('Role'),'Genres'] = 'Adventure'
gamescopy.loc[gamescopy['Genres'].str.contains('Family'),'Genres'] = 'Family'
gamescopy.loc[gamescopy['Genres'].str.contains('Education'),'Genres'] = 'Family'
```

Ví dụ:

Quan sát chính:
Phần lớn các trò chơi có số ngày cập nhật gần nhất thấp (tức là được cập nhật thường xuyên) có nhiều lượt đánh giá hơn. Khi số ngày cập nhật tăng (tức là game ít được cập nhật), số lượng đánh giá có xu hướng giảm đáng kể. Có một số ngoại lệ: Một số game dù ít cập nhật nhưng vẫn có nhiều đánh giá, có thể là các game rất phổ biến. Mặc dù biểu đồ không hiển thị tổng số lượt tải xuống, chúng ta có thể giả định rằng số lượng đánh giá có tương quan với số lượt tải (nhiều đánh giá hơn = nhiều người chơi hơn).

Cập nhật thường xuyên giúp duy trì lượng tương tác. Các game có nhiều lượt đánh giá thường có các bản cập nhật gần đây hơn. Điều này có thể do các yếu tố như cải thiện tính năng, sửa lỗi, hoặc chiến lược tiếp thị thu hút người chơi mới.
Ít cập nhật có thể làm giảm sự quan tâm. Khi một game không được cập nhật trong thời gian dài, số lượng đánh giá có xu hướng thấp hơn, có thể do mất đi sự quan tâm của người dùng hoặc giảm khả năng tiếp cận người chơi mới.

```python
selected_genres = ['Puzzle', 'Action', 'Adventure', 'Family']
filtered_games = gamescopy[gamescopy['Genres'].isin(selected_genres)]
gamescopy2= gamescopy.copy()
import statistics
gamescopy2['In_App_Purchases'] = gamescopy2['In_App_Purchases'].str.replace(',','')
for i in (list(gamescopy2.index.values)):
    if gamescopy2['In_App_Purchases'][i] ==0:
        gamescopy2['In_App_Purchases'][i] = 0.01
    elif gamescopy2['In_App_Purchases'][i] !=0:
        A = str(gamescopy2['In_App_Purchases'][i]).split()
        A = [float(i) for i in A]
        gamescopy2['In_App_Purchases'][i] = statistics.mean(A)

vis13 = sns.stripplot(y=gamescopy2['In_App_Purchases'],
                      x = gamescopy2['Price'],
                      data = gamescopy2,
                      hue = filtered_games['Genres'],
                      dodge = True,
                      size = 3)
vis13.set(xlabel='Price in $', ylabel='Average in App Purchase in $')
vis13.set(xlim=(-0.7,12))
```

Game miễn phí chiếm đa số và có mức chi tiêu trong ứng dụng cao, đặc biệt với Action.
Game trả phí ít có giao dịch trong ứng dụng, nhưng một số thể loại như Family, Puzzle, Adventure vẫn có người mua dù giá cao.
Freemium (miễn phí + IAP) là mô hình tối ưu doanh thu, nhất là với Action.
Game Family và Puzzle có thể bán giá cao + IAP để tối ưu lợi nhuận.

# **Chiến Lược Phát Triển Game Tối Ưu**

## **1. Tập Trung vào Game Dung Lượng Nhỏ & Trung Bình**
### **Lý do:**
- Các game có dung lượng dưới **250MB** dễ tiếp cận nhiều thiết bị, đặc biệt là người dùng di động.
- Những game có dung lượng từ **200MB - 400MB** thường đạt rating cao, đảm bảo chất lượng và tối ưu hiệu suất.

### **Chiến lược:**
- Phát triển game dưới **400MB** để phù hợp với nhiều phân khúc người chơi.
- Tối ưu hóa **đồ họa, gameplay** nhằm duy trì trải nghiệm tốt mà không làm game quá nặng.
- Sử dụng công nghệ **nén và tối ưu hóa dữ liệu** (*asset streaming, texture compression*) để giảm dung lượng nhưng vẫn giữ chất lượng cao.

---

## **2. Tận Dụng Xu Hướng Thể Loại Game**
### **Lý do:**
- Thể loại **Action & Adventure** có dung lượng trung bình cao và đang tăng trưởng mạnh.
- **Puzzle & Family** có kích thước nhỏ nhưng đạt rating cao, phù hợp với trẻ em và người chơi casual.

### **Chiến lược:**
- Phát triển game **Action & Adventure** với dung lượng từ **600MB - 800MB** để thu hút game thủ hardcore.
- Tạo các game **Puzzle & Family** dưới **200MB** để tiếp cận thị trường casual và trẻ em.
- Áp dụng mô hình **Hybrid-Casual**, kết hợp gameplay đơn giản nhưng có chiều sâu để tăng sự gắn bó của người chơi.

---

## **3. Cập Nhật Thường Xuyên để Duy Trì Người Chơi**
### **Lý do:**
- Các game được cập nhật thường xuyên có **nhiều lượt đánh giá hơn**, chứng tỏ mức độ tương tác cao.
- **Action game có vòng đời ngắn (~2.4 năm)**, cần liên tục làm mới để duy trì sự quan tâm.

### **Chiến lược:**
- Xây dựng **lộ trình cập nhật định kỳ (mỗi 2-3 tháng)** để giữ chân người chơi.
- Tận dụng **DLC** hoặc **nội dung mở rộng** thay vì phát triển game mới để tối ưu chi phí.
- Tổ chức các **sự kiện trong game** (*seasonal events, battle pass, live ops*) để duy trì tương tác của người chơi.

---

## **4. Áp Dụng Mô Hình Kinh Doanh Freemium + IAP**
### **Lý do:**
- Game **Freemium** chiếm đa số trên thị trường và có tiềm năng doanh thu cao từ giao dịch trong ứng dụng (**In-App Purchase - IAP**).
- Một số thể loại như **Puzzle & Family** có thể kết hợp giữa **trả phí ban đầu và IAP** để tăng doanh thu.

### **Chiến lược:**
- Đối với **Action & Adventure**, áp dụng mô hình **Freemium + IAP** để tối đa hóa doanh thu.
- Với **Puzzle & Family**, kết hợp **trả phí ban đầu + IAP** để cân bằng lợi nhuận và trải nghiệm người chơi.
- Sử dụng **quảng cáo có chọn lọc** (*rewarded ads, không gây khó chịu*) để tăng doanh thu từ game miễn phí.

---

## **5. Tập Trung vào Thị Trường Trẻ Em & Gia Đình**
### **Lý do:**
- Nhóm tuổi **4+ và 9+** có nhiều game thành công và rating ổn định.
- Nhóm **17+** là thị trường ngách nhưng có tiềm năng phát triển game cao cấp.

### **Chiến lược:**
- Đầu tư phát triển game **Family & Puzzle** dành cho nhóm tuổi **4+ và 9+**, kết hợp yếu tố giáo dục để thu hút phụ huynh.
- Phát triển game **nhập vai, hành động cao cấp (>1GB)** cho nhóm **17+**, nhưng cần tối ưu hiệu suất để tránh tình trạng lag ảnh hưởng đến rating.
- Xây dựng hệ thống **kiểm soát nội dung và an toàn** cho trẻ em để tăng độ tin cậy trên cửa hàng ứng dụng.
